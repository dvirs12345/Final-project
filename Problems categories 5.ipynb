```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
pd.set_option('display.max_columns', 11)
excel = pd.read_excel('Classified 3.xlsx', sheet_name='Sheet1')
```

```python
value_counts_for1 = excel['1'].value_counts()
print(value_counts_for1)
print("\n# of patients", excel.shape[0])
```

```python
value_counts_for1 = value_counts_for1[value_counts_for1>=40]
print(value_counts_for1)
```

```python
value_counts_for1.plot.bar()
```
