# Makes new classified 2,3 (with some alterations)

```python
import pandas as pd
import numpy as np

sheetname = "problems_cat_new"
problems_categories_filename = "problems categories_ver2.xlsx"
id_colname = "h_num_hash"
```

```python
# Replace n>1 with n=1
def make_bigger_one(arr):
    for i in range(len(arr[0][1:])):
        if arr[0][i+1] > 1:
            arr[0][i+1] = 1
    return arr
```

```python
# Lowercase all ProbelmText
def lower_problems(new_classified1):
    for index, row in new_classified1.iterrows():
        new_classified1["ProblemText"][index] = str(new_classified1["ProblemText"][index]).lower()
    
    return new_classified1
```

```python
# Create an array with the values to insert into the DataFrame
def make_array(my_groupby, id, new_classified1):
    """Make the binary array for each patient"""
    my_group = my_groupby.get_group(id)

    arr = pd.DataFrame(None, columns=["trauma", "oncologic (cancer)", "cardiovascular", "pulmonary", "kidney", "metabolic",                                       "gastrointestinal", "neurologic", "hematologic", "vascular", "Sepsis"])
    
    
    new_classified1 = lower_problems(new_classified1) 

    counter = 0
    for index, row in my_group.iterrows():
        problem = str(row["ProblemText"]).lower()

        if counter > 0:
            arr = np.add(arr, new_classified1[problem == new_classified1["ProblemText"]].drop("ProblemText", axis=1).drop("subCategories", axis=1))                                             

        else:
            arr = new_classified1[new_classified1["ProblemText"] == problem]
            arr = arr.drop("ProblemText", axis=1).drop("subCategories", axis=1)
            counter+=1

    arr.insert(0, "id", id, True)
    arr = arr.to_numpy()
    # arr = make_bigger_one(arr)

    return arr
```

```python
# Create the DataFrame and build it
problems_categories = pd.read_excel(problems_categories_filename, sheetname)
new_classified1 = pd.read_excel("new classified1.xlsx", "Sheet1") 
df = pd.DataFrame(None, columns=["id", "trauma", "oncologic (cancer)", "cardiovascular", "pulmonary", "kidney", "metabolic", "gastrointestinal", "neurologic", "hematologic", "vascular", "Sepsis"])

# problems_categories = problems_categories[problems_categories['ProblemText'] != "ד"]
# problems_categories = problems_categories[problems_categories['ProblemText'] != "fracture of"]
# problems_categories = problems_categories[problems_categories['ProblemText'] != "19/03/2015"]
my_groupby = problems_categories.groupby(by=id_colname)
df["id"] = problems_categories[id_colname].unique()

for i in df['id']:
    try:
        df.iloc[(df[df['id'] == i]).index[0]] = make_array(my_groupby, i, new_classified1)
    except Exception:
        print(i)


df.to_excel(r'new Classified2.xlsx', index=False)

print(df)
```
